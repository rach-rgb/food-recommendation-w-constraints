### Constraint 1 Sample

```python
ingr_const = ld.load_ingr_const()
ingr_const.head()
```

```python
const1_pair = []
for u in range (0, user_max):
    for ingr in ingr_const.loc[u].include:
        const1_pair.append((u, ingr))
```

```python
data1 = {
    'user': [],
    'include': [],
    'const': [],
    'no const': [],
    'const size': [],
    'no const size': []
}

for (u, i) in const1_pair:
    data1['user'].append(u)
    data1['include'].append(i)
    data1['const'].append(bf.top_n_const_1(u, i))
    data1['const size'].append(len(data1['const'][-1]))
    result = []
    for (item, rate) in bf.top_k[str(u)]:
        result.append(item)
    data1['no const'].append(result)
    data1['no const size'].append(len(data1['no const'][-1]))
```

```python
df1 = pd.DataFrame({
    'user': data1['user'],
    'include': data1['include'],
    'item list': data1['const'],
    'len': data1['const size'],
    'without constraint': data1['no const']
})
df1
```

```python
df1.to_csv(result_path + 'BF_const1.csv', index = False)
```

### Constraint 3

```python
preference = bf.top_k
```

```python
preference
```

```python
const3_pair = []
const3_answ = []
for u in range (0, user_max):
    for i, r in preference[str(u)]:
        p = int(random.uniform(0, len(preference[str(u)])))
        answ = int(preference[str(u)][p][0])
        
        cur_nutr = bf.recipe_data.loc[int(i)].nutrition
        answ_nutr = bf.recipe_data.loc[int(answ)].nutrition
        target = []
        for x in range (0, len(cur_nutr)):
            target.append(cur_nutr[x]+answ_nutr[x])
        
        for x in range (0, len(target)):
            var = int(random.uniform(0, 2))-1
            target[x] = target[x] * (1 + var/100)
     
        const3_pair.append((u, (int(i), target)))
        const3_answ.append(answ)
        
        if(len(const3_pair)%3==0): 
            break
```

```python
const3_pair[0:10]
```

```python
const3_answ[0:10]
```

```python
data3 = {
    'user': [],
    'history': [],
    'target': [],
    'const': [],
    'no const': [],
    'const size': [],
    'no const size': []
}

for (u, c) in const3_pair:
    i = c[0]
    t = c[1]
    data3['user'].append(u)
    data3['history'].append(i)
    data3['target'].append(t)
    data3['const'].append(bf.top_n_const_3(u, i, t))
    data3['const size'].append(len(data3['const'][-1]))
    result = []
    for (item, rate) in bf.top_k[str(u)]:
        result.append(item)
    data3['no const'].append(result)
    data3['no const size'].append(len(data3['no const'][-1]))
```

```python
df3 = pd.DataFrame({
    'user': data3['user'],
    'history': data3['history'],
    'target': data3['target'],
    'item list': data3['const'],
    'len': data3['const size'],
    'without constraint': data3['no const'],
})
df3
```

```python
df3.to_csv(result_path + 'BF_const3.csv', index = False)
```
