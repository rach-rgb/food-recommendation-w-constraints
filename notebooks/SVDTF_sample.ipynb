```python
# constants
# file path
input_path = "../data/"
output_path = '../result/'

# user/item max
user_max = 1000
food_max = 5000

# const count
const_count = 3
```

```python
def run_model(ctype, idx):
    # PostRec applies constraint after the rating of each item is predicted
    rec = inter_rec.InterRec(input_path +'reduced_rating_data.csv', input_path + 'recipe_data.csv', 
                           input_path + 'const_' + str(ctype) + '.' + str(idx) + '.csv', 
                            TF_algo.SVDtf())
    rec.get_data()  # get rating, attribute, recipe data
    rec.train()  # train with data
    rec.test()  # predict antisets

    df = rec.get_top_n() # return result as Dataframe
    df.to_csv(output_path + 'InterRec_' + str(ctype) + '.' + str(idx) + '.csv')
    
    return rec, df    
```

### Constraint 1

```python
for i in range (0, const_count):
    rec, df = run_model(1, i+1)
    print("Done: ", i + 1)
```
